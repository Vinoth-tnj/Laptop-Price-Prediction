```python
df2=df1.drop(['Model Name','Operating System','Operating System Version','Weight'],axis='columns')
df2.shape
```

```python
df2.head()
```

```python
df2.isnull().sum()
```

```python
df3 = df2.dropna()
df3.isnull().sum()
```

```python
df3['Manufacturer'].unique()
```

```python
df3.columns
```

```python
df3['Screen Size'] = df3['Screen Size'].apply(lambda x: float(x.split(' ')[0]))
```

```python
df3['RAM'] = df3['RAM'].apply(lambda x: int(x.split(' ')[0]))
```

```python
df3.head()
```

```python
df3.Manufacturer = df3.Manufacturer.apply(lambda x: x.strip())

Manufacturer_stats = df3.groupby('Manufacturer')['Manufacturer'].agg('count').sort_values(ascending=False)
Manufacturer_stats
```

```python
dummies = pd.get_dummies(df3.Manufacturer)
dummies.head(3)
```

```python
df4 = pd.concat([df3,dummies],axis='columns')
df4.head()
```

```python
df4.shape
```

```python
dummies = pd.get_dummies(df4.Category)
dummies.head(3)
```

```python
df5 = pd.concat([df4,dummies],axis='columns')
df5.head()
```

```python
df5['Screen'].unique()
```

```python
dummies = pd.get_dummies(df5.Screen)
dummies.head(3)
```

```python
df6 = pd.concat([df5,dummies],axis='columns')
df6.head()
```

```python
dummies = pd.get_dummies(df6.CPU)
dummies.head(3)
```

```python
df7 = pd.concat([df6,dummies],axis='columns')
df7.head()
```

```python
dummies = pd.get_dummies(df7.GPU)
dummies.head(3)
```

```python
df8 = pd.concat([df7,dummies],axis='columns')
df8.head()
```

```python
dummies = pd.get_dummies(df8.Storage)
dummies.head(3)
```

```python
df9 = pd.concat([df8,dummies],axis='columns')
df9.head()
```

```python
df10 = df9.drop(['Manufacturer','Category','Screen','CPU','Storage','GPU'],axis='columns')
df10.shape
```

```python
df10.head()
```

```python
X = df10.drop(['Price'],axis='columns')
X.head(3)
```

```python
y = df10.Price
y.head(3)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import Lasso
lr_clf = Lasso()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(Lasso(), X, y, cv=cv)
```

```python
def predict_price(screen_size,ram,manufaturer,category,screen,cpu,storage,gpu):    
    manufaturer_index = np.where(X.columns==manufaturer)[0][0]
    category_index = np.where(X.columns==category)[0][0]
    screen_index = np.where(X.columns==screen)[0][0]
    cpu_index = np.where(X.columns==cpu)[0][0]
    storage_index = np.where(X.columns==storage)[0][0]
    gpu_index = np.where(X.columns==gpu)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = screen_size
    x[1] = ram
    if manufaturer_index >= 0:
        x[manufaturer_index] = 1
    if category_index >= 0:
        x[category_index] = 1
    if screen_index >= 0:
        x[screen_index] = 1
    if cpu_index >= 0:
        x[cpu_index] = 1
    if storage_index >= 0:
        x[storage_index] = 1
    if gpu_index >= 0:
        x[gpu_index] = 1

    return lr_clf.predict([x])[0]
```

```python
predict_price(15.6,8,'HP','Notebook','Full HD 1920x1080','Intel Core i5 7200U 2.5GHz','256 GB SSD','Intel HD Graphics 620')
```
